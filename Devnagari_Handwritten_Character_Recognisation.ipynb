```python
import numpy as np
```

```python
from concurrent.futures import  ThreadPoolExecutor
```

```python
def image_flattening(img_np_array):
  return img_np_array.reshape(img_np_array.shape[0]*img_np_array.shape[1],)
```

```python
def read_image(img_path):
  return plt.imread(img_path)
```

```python
def single_dir_img_path_list(base_train_dir_path):
  img_list = os.listdir(base_train_dir_path)
  with ThreadPoolExecutor(max_workers=4) as pool:
    img_path_list = list(pool.map(lambda x: base_train_dir_path+"/"+x,img_list))
  return img_path_list
```

```python
dir_path_list = single_dir_img_path_list(base_train_path)
```

```python
dir_path_list
```

```python
len(dir_path_list)
```

```python
img_path_list = single_dir_img_path_list(dir_path_list[0])
```

```python
len(img_path_list)
```

```python
with ThreadPoolExecutor(max_workers=4) as pool:
  read_imgs_list = list(pool.map(read_image,img_path_list))
```

```python
len(read_imgs_list)
```

```python
read_imgs_list[0].shape
```

```python
def single_dir_imgs_array_list(img_path_list):
  with ThreadPoolExecutor(max_workers=4) as pool:
    imgs_np_array_list = list(pool.map(read_image, img_path_list))
  return imgs_np_array_list
```

```python
imgs_np_array_list = single_dir_imgs_array_list(img_path_list)
```

```python
len(imgs_np_array_list)
```

```python
imgs_np_array_list[0].shape
```

```python
def single_dir_imgs_matrix(imgs_np_array_list):
  with ThreadPoolExecutor(max_workers=4) as pool:
    reshaped_imgs_list = list(pool.map(image_flattening, imgs_np_array_list))
  return np.array(reshaped_imgs_list)
```

```python
reshaped_imgs_list = single_dir_imgs_matrix(imgs_np_array_list)
```

```python
reshaped_imgs_list
```

```python
reshaped_imgs_list.shape
```

```python
reshaped_imgs_matrix_list = list()
for single_dir_path in dir_path_list:
  img_path_list = single_dir_img_path_list(single_dir_path)
  imgs_np_array_list = single_dir_imgs_array_list(img_path_list)
  reshaped_imgs_list = single_dir_imgs_matrix(imgs_np_array_list)
  reshaped_imgs_matrix_list.append(reshaped_imgs_list)
```

```python
training_data_input_features = np.concatenate(reshaped_imgs_matrix_list)
```

```python
### Accuracy on both training as well as testing data is very low. Therefore, need to apply PCA on our model.
```

```python
### Importing PCA from sklearn library.
from sklearn.decomposition import PCA
### Also, Importing some crucial scores from sklearn to check on how many PCA n_components our model is working best.
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
num_pixels = np.arange(100,1024,50)
```

```python
num_pixels
```

```python
### Now, applying PCA :-
```

```python
def fetch_performance_metrics(num_components):
    pca = PCA(n_components=num_components)
    reduced_training_features = pca.fit_transform(training_features)
    reduced_testing_features = pca.transform(testing_features)
    model = GaussianNB()
    model.fit(reduced_training_features, training_labels)
    training_predictions = model.predict(reduced_training_features)
    testing_predictions = model.predict(reduced_testing_features)
    training_accuracy = accuracy_score(training_labels, training_predictions)
    training_precision = precision_score(training_labels, training_predictions, average='weighted')
    training_recall = recall_score(training_labels, training_predictions, average='weighted')
    training_f1_score = f1_score(training_labels, training_predictions, average='weighted')
    testing_accuracy = accuracy_score(testing_labels, testing_predictions)
    testing_precision = precision_score(testing_labels, testing_predictions, average='weighted')
    testing_recall = recall_score(testing_labels, testing_predictions, average='weighted')
    testing_f1_score = f1_score(testing_labels, testing_predictions, average='weighted')
    print("Fetch performance metrics for",num_components,"Principal Components")
    my_dict = {
    num_components: {
        'Train': {
            'F1 Score': training_f1_score,
            'Precision': training_precision,
            'Recall': training_recall,
            'Accuracy': training_accuracy
        },
        'Test': {
            'F1 Score': testing_f1_score,
            'Precision': testing_precision,
            'Recall': testing_recall,
            'Accuracy': testing_accuracy
        }
    }},
    return (my_dict)
```

```python
complete_report = dict()

for num_components in num_pixels:

  complete_report[num_components] = fetch_performance_metrics(num_components)
```

```python
complete_report
```

```python
### PCA will be best for 150 n_components for this specific project as the accuracy of our model after applying the PCA for 150 n_components is near to 68% which is highest than any other accuracy.
```
